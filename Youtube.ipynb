```python
import pandas as pd
from datetime import datetime
import json
import glob
```

```python
appended_data = []
for filename in glob.glob('./data/*videos.csv'):
    country_name = filename.split('/')[2][:2]
    print(filename)
    print(country_name)
    df = pd.read_csv('./data/'+country_name+'videos.csv',encoding = "ISO-8859-1")
    df['country'] = country_name
    df['trending_date'] = pd.to_datetime(df['trending_date'], format='%y.%d.%m')
    df['publish_time'] = pd.to_datetime(df['publish_time'], format='%Y-%m-%dT%H:%M:%S.%fZ').dt.date
    category_id = ['category_id']
    for column in category_id:
        df[column] = df[column].astype(str)
    id_to_category = {}

    with open('./data/'+country_name+'_category_id.json', 'r') as f:
        data = json.load(f)
        for category in data['items']:
            id_to_category[category['id']] = category['snippet']['title']

    df.insert(4, 'Category Name', df['category_id'].map(id_to_category))
    df = df.rename(columns = {'Category':'Category Name'})
    df = df.dropna(subset=['Category Name'])
    appended_data.append(df)
    
appended_data = pd.concat(appended_data)
```
